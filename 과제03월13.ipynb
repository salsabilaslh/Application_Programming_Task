```python
%%writefile my_utils.py

# variable
E = 2.71828

# function
def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

# class
class TemperatureConverter:

    def celsius_to_kelvin(self, c):
        return c + 273.15

    def kelvin_to_celsius(self, k):
        return k - 273.15
```
